# 8장. 오차 역전파

오차 역전파는 4장에서 배운 경사 하강법의 확장 개념임

## 1. 오차 역전파의 개념

< 가중치를 구하는 방법 > <br>
임의의 가중치를 선언하고 결괏값을 이용해 오차를 구한 뒤 이 오차가 최소인 지점 으로 계속해서 조금씩 이동시킴. 이 오차가 최소가 되는 점(미분했을 때 기울 기가 0이 되는 지점)이 가중치

image.png

오차 역전파: 다층 퍼셉트론에서의 최적화 과정

< 오차 역전파 구동 방식 > <br>
1. 임의의 초기 가중치(W)를 준 뒤 결과(y out)를 계산한다. <br>
2. 계산 결과와 우리가 원하는 값 사이의 오차를 구한다. <br>
3. 경사 하강법을 이용해 바로 앞 가중치를 오차가 작아지는 방향으로 업데이트 한다. <br>
4. 위 과정을 더이상 오차가 줄어들지 않을 때까지 반복한다.

오차가 작아지는 방향으로 업데이트한다. <br>
= 미분 값이 0에 가까워지는 방향으로 나아간다. <br>
= 기울기가 0이 되는 방향으로 나아간다. <br>
= 가중치에서 기울기를 뺐을 때 가중치의 변화가 전혀 없는 상태
