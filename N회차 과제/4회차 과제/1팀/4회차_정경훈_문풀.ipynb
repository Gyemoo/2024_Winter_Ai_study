### 문제 1
seaborn 라이브러리에 내장된 'penguins' 데이터셋을 불러와
heatmap 합수를 사용하여 펭귄 데이터셋의 상관도를 시각화 하고
펭귄의 몸무게와 가장 상관관계가 강한 항목이 무엇인지 말하시요
(그래프의 크기는 4 x 4 이다.)

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('penguins')
correlation_matrix = df.corr()
plt.figure(figsize = (4, 4))
sns.heatmap(correlation_matrix, annot = True)

plt.show()
```

```python
filtered_corr = correlation_matrix.loc[correlation_matrix.index != 'body_mass_g','body_mass_g']

max_corr_variable = filtered_corr.idxmax()
max_corr_value = filtered_corr.max()

max_corr_variable,max_corr_value
```

펭귄의 팔 길이(flipper_length_mm) (0.87)
