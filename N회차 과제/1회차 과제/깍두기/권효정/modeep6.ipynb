# 6장. 퍼셉트론

뉴런의 메커니즘이 로지스틱 회귀와 많이 닮았음 <br>
입력 값을 놓고 활성화 함수에 의해 일정한 수준을 넘으면 참을, 그렇지 않으면 거짓을 내보내는 이 간단한 회로가 하는 일이 뉴런과 같음.

신경망의 기본 구조: 뉴런과 뉴런이 서로 새로운 연결을 만들기도 하고, 필요에 따라 위치를 바꾸는 것처럼, 여러 층의 퍼셉트론을 서로 연결시키고 복잡하게 조합하여 입력 값에 대한 판단을 하게 하는 것

## 1. 가중치, 가중합, 바이어스, 활성화 함수

y= ax + b (a는 기울기, b는 y절편) <br>
→ y= wx + b (w는 가중치, b는 바이어스)

y=wx+b에서의 b는 '편향, 선입견'이라는 뜻인 바이어스에서 따온 b임.

- 가중합: 입력 값(x)과 가중치(w)의 곱을 모두 더한 다음 거기에 바이어스(b)를 더한 값. <br>
- 활성화 함수: 0과 1을 판단하는 함수

# 2. 퍼셉트론의 과제

단 하나의 퍼셉트론으로는 많은 것을 기대할 수가 없음. <br>
선형 회귀와 로지스틱 회귀를 통해 머신러닝이 결국 선이나 2차원 평면을 그리는 작업이라는 것을 알 수 있음 <br>
퍼셉트론 역시 선을 긋는 작업임

## 3.XOR 문제

게이트: 0과 1을 입력해 하나의 값을 출력하는 회로

AND 게이트: x1와 x2 둘 다 1일 때 결과값이 1로 출력됨. <br>
OR 게이트: 둘 중 하나라도 1이면 결과값이 1로 출력됨. <br>
XOR 게이트: 둘 중 하나만 1일 때 1이 출력됨.

image.png

결과값이 0이면 흰점, 1이면 검은점 image.png

AND 게이트와 OR 게이트는 직선을 그어 결과값이 1인 값(검은점)을 구별 가능 <br>
그러나 XOR 게이트는 선을 그어 구분 불가

<br>
Marvin Minsky 교수의 논문 <퍼셉트론즈>에 나온 내용 <br>
' 뉴런 → 신경망 → 지능 ' 이라는 도식을 따라 <br>
' 퍼셉트론 → 인공 신경망 → 인공지능 ' 이 가능할 것이라 생각함. <br>

이 과정은 쉽지 않았음 (간단한 XOR 문제도 해결 불가) <br>
이 논문 이후 인공지능 연구가 한동안 침체기를 겪음 <br>
10여 년 후 ' 다층 퍼셉트론 '이 이 문제를 해결하게 됨.
