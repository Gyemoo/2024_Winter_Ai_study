### 문제 <br>
Q. 아래와 같이 수정한 데이터 값을 바탕으로 기울기와 y 절편을 구해 일차 함수 식을 도출해보시오. <최소 제곱법 공식 사용, 코드를 이용해 구현 (나누어 떨어집니다)> <br>
image.png

```python
import numpy as np
```

```python
x = [1,3,5]
y = [68,89,98]
```

```python
mx=np.mean(x)
my=np.mean(y)

divisor=sum([(i-mx)**2 for i in x])

def study(x,mx,y,my):
  d=0
  for i in range(len(x)):
    d+=(x[i]-mx)*(y[i]-my)
  return d
dividend=study(x,mx,y,my)

a=dividend/divisor
b=my-(mx*a)

print("y = %dx + %d" %(a,b))
```

### 문제 <br>
data = [[3, 78], [5, 85], [7, 89], [9, 94]] <br>
lr = 0.02 <br>
epochs = 3001 <br>

위 조건을 이용해 구한 기울기와 절편을 이용한 그래프를 출력하라.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=[[3,78],[5,85],[7,89],[9,94]]
lr=0.02
epochs=3001
```

```python
x=[i[0] for i in data]
y=[i[1] for i in data]

ax=np.array(x)
ay=np.array(y)

a=0
b=0

for i in range(epochs):
  y_pred=a*ax+b
  error=ay-y_pred
  a_diff=-(2/len(ax))*sum(ax*(error))
  b_diff=-(2/len(ax))*sum(error)

  a=a-lr*a_diff
  b=b-lr*b_diff

y_pred=a*ax+b
plt.scatter(x,y)
plt.plot([min(ax),max(ax)],[min(y_pred),max(y_pred)])
plt.show()
```
